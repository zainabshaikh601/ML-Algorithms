```python
# Importing necessary packages
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('/content/Income.csv')
df.head()
```

```python
# Visualize the initial data
plt.scatter(df.Age, df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
# Apply k-means clustering
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
# Add cluster labels to the dataset
df['cluster']=y_predicted
df.head()
```

```python
# Calculating centeroid
km.cluster_centers_
```

```python
# Visualize the clustered data
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'], color='green')
plt.scatter(df2.Age,df2['Income($)'], color='red')
plt.scatter(df3.Age,df3['Income($)'], color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centeroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
# Scale the features between 0 and 1

scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] =  scaler.transform(df[['Age']])
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
```

```python
# Apply k-means clustering after scaling
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
# Add cluster labels to the dataset
df['cluster']=y_predicted
df.head()
```

```python
# Calculating centeroid
km.cluster_centers_
```

```python
# Visualize the clustered data after scaling
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'], color='green')
plt.scatter(df2.Age,df2['Income($)'], color='red')
plt.scatter(df3.Age,df3['Income($)'], color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centeroid')
plt.legend()
```

```python
# Calculate the sum of squared errors (SSE) for different values of k
sse = []
k_rng = range(1,10)
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
# Plot SSE versus the number of clusters (k)
plt.xlabel('K')
plt.ylabel('Sum of squarred error')
plt.plot(k_rng,sse)
```
